# 使用するカタログとスキーマを指定

```sql
USE CATALOG handson;
USE SCHEMA sports_orders;
```

# ブロンズテーブル初期作成（最初の1回のみ）

```sql
CREATE TABLE IF NOT EXISTS bronze_sports_orders (
  order_id INT,
  customer_name STRING,
  region STRING,
  order_date DATE,
  product STRING,
  category STRING,
  quantity INT,
  price DOUBLE
);
```

# 新しいCSVをブロンズにCOPY INTO（CDC反映）

```sql
COPY INTO bronze_sports_orders
  FROM (SELECT order_id, customer_name, region, order_date, product, category, quantity, price
          FROM '/Volumes/handson/sports_orders/csv_files')
  FILEFORMAT = CSV
  PATTERN = 'sports_orders*\.csv'
  FORMAT_OPTIONS ('inferSchema' = 'true', 'mergeSchema' = 'true', 'header' = 'true')
  COPY_OPTIONS ('mergeSchema' = 'true');
```
